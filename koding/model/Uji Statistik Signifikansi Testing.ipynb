# Skenario Pemrosesan Augmentasi Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import friedmanchisquare
import scikit_posthocs as sp
import os

print(os.getcwd())

# Load file CSV
data = r"C:\Users\ASUS\OneDrive - Duta Wacana Christian University\Documents\8th semester\Skripsi\koding\model\hasil_skenario_aug.csv"
df = pd.read_csv(data)

# Cek isi data
print(df.head())

data_values = df.values

# Uji Friedman (uji non-parametrik untuk data berulang)
stat, p = friedmanchisquare(*[data_values[:, i] for i in range(data_values.shape[1])])
print(f"Hasil Uji Friedman:\nStatistik: {stat:.4f}, p-value: {p:.4f}")

# Uji lanjut (post-hoc) jika p-value < 0.05
if p < 0.05:
    print("Terdapat perbedaan signifikan. Lanjut ke uji post-hoc Nemenyi...")
    p_values_matrix = sp.posthoc_nemenyi_friedman(data_values)

    # Visualisasi heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(p_values_matrix, annot=True, fmt=".3f", cmap="coolwarm", xticklabels=df.columns, yticklabels=df.columns)
    plt.title("Heatmap P-Value Uji Nemenyi (Posthoc Friedman Test)")
    plt.show()
else:
    print("Tidak ada perbedaan signifikan antara metode augmentasi.")
```
